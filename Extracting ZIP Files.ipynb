```python
import json
import requests
import os
from datetime import datetime
import zipfile
import tempfile
from collections import defaultdict
import time
import shutil
```

```python
zip_file_folder = './snapchat_downloads_renamed'
```

```python
delete_dir = os.path.join(zip_file_folder, "extracted_zips_to_delete_later")
os.makedirs(delete_dir, exist_ok=True)
```

```python
for filename in os.listdir(zip_file_folder):
    
    if not filename.lower().endswith(".zip"):
        continue
        
    zip_path = zip_file_folder + "/" + filename
    base_name = filename.split(".")[0]
    
    with zipfile.ZipFile(zip_path, "r") as z:
        
        file_counter = 1
        
        for member in z.infolist():
            
            original_file_name = member.filename.split(".")[0]
            original_file_extension = member.filename.split(".")[-1]

            if original_file_name.endswith("-main"):
                new_name = f"{base_name}_{file_counter}_main.{original_file_extension}"

            elif original_file_name.endswith("-overlay"):
                new_name = f"{base_name}_{file_counter}_overlay.{original_file_extension}"

            else:
                new_name = f"{base_name}_{file_counter}.{original_file_extension}"
                
            out_path = zip_file_folder + "/" + new_name
            
            
            with z.open(member) as source, open(out_path, "wb") as target:
                shutil.copyfileobj(source, target)
                print("  Saved:", new_name)
                file_counter += 1

    shutil.move(zip_path, os.path.join(delete_dir, filename))
    print("Moved zip to delete folder:", filename)
```
